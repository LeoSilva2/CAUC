```python
import pandas as pd
pd.set_option("display.max_colwidth", 150)
```

```python
import csv 
import numpy
import matplotlib.pyplot as plt
import datetime as dt
```

```python
def cauc():
  url = "https://www.tesourotransparente.gov.br/ckan/dataset/72b5f371-0c35-4613-8076-c99c821a6410/resource/07af297a-5e59-494a-a88a-55ddfd2f4b01/download/Relatorio-Situacao-de-Varios-Entes---Municipios---UF-Todas---Abrangencia-1.csv"
  return pd.read_csv(url, sep=';',encoding='ISO-8859-1', header=3)
```

```python
cauc2 = cauc()



#Excluindo colunas
cauc2 = cauc2.dropna(how='all', axis=1)

#Excluindo linhas
cauc2 = cauc2.dropna(how='any', axis=0)

cauc2

qtd_total_cidades= cauc2['UF'].value_counts()
qtd_total_cidades
```

Filtra todos os estados do Brasil

```python
dados_cidade = cauc2.loc[(cauc2['1.1'] != '!') & (cauc2['1.3'] != '!') & (cauc2['1.4'] != '!') & (cauc2['1.5'] != '!') & (cauc2['2.1.1'] != '!') & (cauc2['2.1.2'] != '!') & (cauc2['3.1.1'] != '!') & (cauc2['3.1.2'] != '!') & (cauc2['3.2.1'] != '!') & (cauc2['3.2.2'] != '!') & (cauc2['3.2.3'] != '!') & (cauc2['3.2.4'] != '!') & (cauc2['3.3'] != '!') & (cauc2['3.4'] != '!')
 & (cauc2['3.5'] != '!') & (cauc2['4.1'] != '!') & (cauc2['4.2'] != '!') & (cauc2['5.1'] != '!') & (cauc2['5.2'] != '!') & (cauc2['5.3'] != '!') & (cauc2['5.4'] != '!') ]
dados_cidade
```

```python
#Excluindo colunas
dados_cidade = dados_cidade.dropna(how='all', axis=1)


#Excluindo linhas
dados_cidade = dados_cidade.dropna(how='any', axis=0)



arquivo1 = 'CAUC_BRASIL.xlsx'
arquivo1
dados_cidade.to_excel(arquivo1)
```

```python
##Gerando Grafico 

#Qtd_OK=pd.DataFrame({'UF': dados_cidade['UF'].value_counts(),
                    # 'Cidades': qtd_total_cidades})

                    
Qtd_OK= dados_cidade['UF'].value_counts() 
Qtd_OK 
```

```python
##Gerando Grafico 
#Qtd_OK = Qtd_OK.sort_values('UF', ascending=False)
Qtd_OK=pd.DataFrame({'UF': dados_cidade['UF'].value_counts(),
                     'Cidades': qtd_total_cidades} 
                    
                    )
Qtd_OK = Qtd_OK.reset_index()

Qtd_OK = Qtd_OK.sort_values('UF', ascending=False)
Qtd_OK
```

```python
Qtd_OK['Percentual'] = Qtd_OK['UF'] / Qtd_OK['Cidades'] * 100
Qtd_OK = Qtd_OK.sort_values('UF', ascending=False)
#Qtd_OK.reset_index(inplace=True)
Qtd_OK
```

```python



plt.figure(figsize=(20,8))
plt.title('Quantidade de Cidades por UF com CAUC OK')
plt.ylabel('Quantidade de cidades com CAUC OK')

dia =(dt.datetime.now()).day
mes =(dt.datetime.now()).month
ano =(dt.datetime.now()).year
dataformatada = dia, mes, ano
plt.xlabel(dataformatada)

plt.bar(Qtd_OK['index'],Qtd_OK['UF'])

plt.savefig('/content/qtd_cidades_cauc_ok.png')

```
